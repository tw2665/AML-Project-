```
## setup

from __future__ import absolute_import, division, print_function

import base64
import imageio
import IPython
import matplotlib
import matplotlib.pyplot as plt
import PIL.Image
import pyvirtualdisplay

import tensorflow as tf

from tf_agents.agents.dqn import dqn_agent
from tf_agents.agents.ddpg import critic_network
from tf_agents.agents.sac import sac_agent
from tf_agents.drivers import dynamic_step_driver
from tf_agents.environments import suite_gym
from tf_agents.environments import tf_py_environment
from tf_agents.eval import metric_utils
from tf_agents.metrics import tf_metrics
from tf_agents.networks import q_network
from tf_agents.networks import normal_projection_network
from tf_agents.networks import actor_distribution_network
from tf_agents.policies import random_tf_policy
from tf_agents.policies import greedy_policy
from tf_agents.replay_buffers import tf_uniform_replay_buffer
from tf_agents.trajectories import trajectory
from tf_agents.utils import common

tf.compat.v1.enable_v2_behavior()

# Set up a virtual display for rendering OpenAI gym environments.
display = pyvirtualdisplay.Display(visible=0, size=(1400, 900)).start()

# clear rendered env display
from IPython.display import clear_output
# freeze rendered env display
from time import sleep
```

```
## hyperparameter

# CartPole has discrete action space, which is not supported by SAC
# ENV_NAME = "CartPole-v0"
# continuous mountain has float32 action spec
# negative is to push to left, positive is to push to right
ENV_NAME = "MountainCarContinuous-v0"
critic_joint_fc_layer_params = (256, 256)
actor_fc_layer_params = (256, 256)
global_step = tf.compat.v1.train.get_or_create_global_step()
actor_learning_rate = 3e-4
critic_learning_rate = 3e-4 
alpha_learning_rate = 3e-4 
target_update_tau = 0.005
target_update_period = 1
loss_fn = tf.compat.v1.losses.mean_squared_error
gamma = 0.99
reward_scale_factor = 1.0
gradient_clipping = None
num_eval_episodes = 30
replay_buffer_capacity = 1000000
initial_collect_steps = 10000
batch_size = 256
collect_steps_per_iteration = 1
# number of iteration in each training, not episode
# num_iterations = 1000000
num_iterations = 100000
# interval to show loss in training
log_interval = 5000
# interval to show average return in evaluation
eval_interval = 10000
```

```
## environment

#@test {"skip": true}
train_py_env = suite_gym.load(ENV_NAME)
eval_py_env = suite_gym.load(ENV_NAME)
train_env = tf_py_environment.TFPyEnvironment(train_py_env)
eval_env = tf_py_environment.TFPyEnvironment(eval_py_env)

train_py_env.reset()
PIL.Image.fromarray(train_py_env.render())

# print("observation", env.time_step_spec().observation)
# print("action", env.action_spec())
# print("reward", env.time_step_spec().reward)
```

```
## network

observation_spec = train_env.observation_spec()
action_spec = train_env.action_spec()

# used in actor net
def normal_projection_net(action_spec, init_means_output_factor = 0.1):
  return normal_projection_network.NormalProjectionNetwork(
      action_spec,
      mean_transform = None,
      state_dependent_std = None,
      init_means_output_factor = init_means_output_factor,
      std_transform = sac_agent.std_clip_transform,
      scale_distribution = True
  )

# critic net
critic_net = critic_network.CriticNetwork((observation_spec, action_spec),
                                          observation_fc_layer_params = None,
                                          action_fc_layer_params = None,
                                          joint_fc_layer_params = critic_joint_fc_layer_params)

# actor net
actor_net = actor_distribution_network.ActorDistributionNetwork(
    observation_spec,
    action_spec,
    fc_layer_params = actor_fc_layer_params,
    continuous_projection_net = normal_projection_net
)
```

```
## agent

tf_agent = sac_agent.SacAgent(
    train_env.time_step_spec(),
    action_spec,
    actor_network = actor_net,
    critic_network = critic_net,
    actor_optimizer =  tf.compat.v1.train.AdamOptimizer(learning_rate = actor_learning_rate),
    critic_optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate = critic_learning_rate),
    alpha_optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate = alpha_learning_rate),
    target_update_tau = target_update_tau,
    target_update_period = target_update_period,
    td_errors_loss_fn = loss_fn,
    gamma = gamma,
    reward_scale_factor = reward_scale_factor,
    gradient_clipping = gradient_clipping,
    train_step_counter = global_step
)

tf_agent.initialize()
```

```
## policy

eval_policy = greedy_policy.GreedyPolicy(tf_agent.policy)
collect_policy = tf_agent.collect_policy
```

```
## evaluation

def compute_avg_return(environment, policy, num_episodes = 5):
  # initialize for this function
  total_return = 0.0
  for _ in range(num_episodes):
    # initialize for per episode
    time_step = environment.reset()
    episode_return = 0.0
    while not time_step.is_last():
      action_step = policy.action(time_step)
      time_step = environment.step(action_step.action)
      episode_return += time_step.reward

    # debug
    print(episode_return)
    
    # total return is accumulated returns from each episode
    total_return += episode_return

  avg_return = total_return / num_episodes
  return avg_return.numpy()[0]
```

```
## replay buffer

replay_buffer = tf_uniform_replay_buffer.TFUniformReplayBuffer(
    data_spec = tf_agent.collect_data_spec,
    batch_size = train_env.batch_size,
    max_length = replay_buffer_capacity
)
```

```
# driver

initial_collect_driver = dynamic_step_driver.DynamicStepDriver(
    train_env,
    collect_policy,
    observers = [replay_buffer.add_batch],
    num_steps = initial_collect_steps
)
```

```
# data collection

initial_collect_driver.run()
```

```
# dataset for training

dataset = replay_buffer.as_dataset(
    # for optimization
    num_parallel_calls = 3,
    # batch size for each training
    sample_batch_size = batch_size,
    # SAC agent needs current and next observation to compute loss, so sample two adjacent rows
    num_steps = 2
# for optimization (https://www.tensorflow.org/guide/data_performance)
).prefetch(3)

# make an iterator for training
iterator = iter(dataset)
```

```
# driver 2 (What's the difference of initial collect driver and collect driver)

collect_driver = dynamic_step_driver.DynamicStepDriver(
    train_env,
    collect_policy,
    observers = [replay_buffer.add_batch],
    num_steps = collect_steps_per_iteration
)
```

```
# training

# what's this?
#@test {"skip": true}
%%time

# for optimization
tf_agent.train = common.function(tf_agent.train)
collect_driver.run = common.function(collect_driver.run)

# reset train step
tf_agent.train_step_counter.assign(0)

# initial evaluation before training (we can check how we improve)
# eval_policy is object of tf_agent.policy with only applying greedy algorithm, 
# that is only exploitation because we done training
# at this point, eval_policy is not trained yet
avg_return = compute_avg_return(eval_env, eval_policy, num_eval_episodes)
# We will accumulate return in evaluation environment and policy through training
returns = [avg_return]
# additionally we store loss
losses = []

# start training
for _ in range(num_iterations):

  # I don't know why we need to run collect_driver
  for _ in range(collect_steps_per_iteration):
    collect_driver.run()

  # sample training dataset
  experience, unused_info = next(iterator)
  # .train returns total loss
  # In SAC, total loss = critic loss + actor loss + alpha loss
  train_loss = tf_agent.train(experience)
  # return in which iteration we are in
  step = tf_agent.train_step_counter.numpy()

  # show logs
  if step % log_interval == 0:
    # .loss contains the above total loss
    print('step = {0}: loss = {1}'.format(step, train_loss.loss))
    losses.append(train_loss.loss)
    
  if step % eval_interval == 0:
    # eval_policy is updated along training
    avg_return = compute_avg_return(eval_env, eval_policy, num_eval_episodes)
    print('step = {0}: Average return = {1}'.format(step, avg_return))
    returns.append(avg_return)
```
